### Selecting movie_id,'title','overview','genres','keywords','cast','crew' columns for content based recommender 
***movie_id is required for images\

```python
movies_2=movies_1[['movie_id','title','overview','genres','keywords','cast','crew' ]]
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies_2['genres'] = movies_2['genres'].apply(convert)
movies_2.head()
```

```python
type(movies_2.loc[(0,'keywords')])
```

```python
movies_2['keywords'] = movies_2['keywords'].apply(convert)
movies_2.head()
```

```python
movies_2.cast[0]
```

**we will only select top3 names from the cast**

```python
movies_2['cast']=movies_2['cast'].apply(convert).apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies_2.loc[:,'crew']=movies_2['crew'].apply(fetch_director)
```

```python
movies_2.head(1)
```

```python
#function to remove spaces 
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies_2.loc[:,'cast'] = movies_2['cast'].apply(collapse)
movies_2.loc[:,'crew'] = movies_2['crew'].apply(collapse)
movies_2.loc[:,'genres'] = movies_2['genres'].apply(collapse)
movies_2.loc[:,'keywords'] = movies_2['keywords'].apply(collapse)
```

```python
movies_2.head()
```

```python
movies_2['overview'] = movies_2['overview'].apply(lambda x:x.split())
```

```python
#combining all the columns in to tags
movies_2['tags'] = movies_2['overview'] + movies_2['genres'] + movies_2['keywords'] + movies_2['cast'] + movies_2['crew']
```

```python
movies_2.head(1)
```

```python
movies_final=movies_2.drop(columns=['overview','genres','keywords','cast','crew'])
movies_final.head(1)
```

```python
movies_final['tags'] = movies_final['tags'].apply(lambda x: " ".join(x))
movies_final.head()
```

### We need to convert text to vectors
#### we will simply use bag of words 
we will select the top 5000 most common words

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
from nltk.corpus import stopwords
import re
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
text1=movies_final['tags'][0]
```

```python
stem(text1)
```

```python
movies_final['tags']=movies_final['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(movies_final['tags']).toarray().shape
```

```python

vectors=cv.fit_transform(movies_final['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

#### 4806 movies finding distances between 4806 movies

```python
similarity[0]
```

## score of 1st movie with itself is 1

```python
#Recommend top 5 movies
# we need to sort without losing index
def recommend(movie):
    index = movies_final[movies_final['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies_final.iloc[i[0]].title)
        
```

```python
recommend('Batman Begins')
```
